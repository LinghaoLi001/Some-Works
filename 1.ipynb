```python
df11=data1
x=df11['demand_mbps_km2'].values
y=df11['Cost.per.km2'].values
x=x.reshape(-1,1)
y=y.reshape(-1,1)
regr11 = linear_model.LinearRegression().fit(x,y)
print ('Coefficients: ', regr11.coef_)
print ('Intercept: ',regr11.intercept_)
plt.figure(figsize=(20,10))
plt.scatter(x, y,  color='blue')
plt.plot(x, regr11.intercept_[0]+ regr11.coef_[0][0]*x, '-r')
plt.xlabel("demand")
plt.ylabel("Cost")
```
